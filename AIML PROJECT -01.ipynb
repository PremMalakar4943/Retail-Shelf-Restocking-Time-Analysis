##### Retail Shelf Restocking Time Analysis #####

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('retail_store_inventory.csv', nrows=100)
```

```python
np.random.seed(42)
df['Restocking Delay'] = np.random.randint(1, 8, size=100)
```

```python
# I am calculating the mean here to see the average wait time
avg_delay = df['Restocking Delay'].mean()
# Standard deviation shows how much the delay changes (risk)
std_dev_delay = df['Restocking Delay'].std()
```

```python
print("Average Delay:", avg_delay)
print("Delay Variation (Std):", std_dev_delay)
```

```python
item_analysis = df.groupby('Category')['Restocking Delay'].mean()
print("\nDelay by Category:\n", item_analysis)
```

```python
x= ["Clothing","Electronics","Furniture","Groceries","Toys"]
h= [381,420,430,386,500]
plt.bar(x,h)
plt.xlabel("Item")
plt.ylabel("Average Delay")
plt.title("Average Delay by Item")
plt.show()
```
